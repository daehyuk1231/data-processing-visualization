```python
import numpy as np
import random

# 5명의 사람의 나이와 키를 랜덤하게 생성하여 2중 리스트 생성
person_list = []
for i in range(5):
    person = [random.randint(28, 50), random.randint(100, 900)]
    person_list.append(person)

# 2중 리스트를 NumPy 2차원 배열로 변환
dataset = np.array(person_list)

print(dataset)
print("배열 형태: {}D, {}".format(dataset.ndim, dataset.shape))
print("샘플 수: {}".format(dataset.shape[0]))
print("특성 수: {}".format(dataset.shape[1]))
```
