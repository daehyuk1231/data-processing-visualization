0차원 배열

```python
import numpy as np

ndarray = np.array(1)

print(f"배열 타입: {type(ndarray)}")
print(f"배열 형태: {ndarray.shape}, {ndarray.ndim}D")
print(f"배열 요소 타입: {ndarray.dtype}")
print(ndarray)
```

1차원 배열

```python
import numpy as np

ndarray = np.array([-1, 2, 7, 19])

print(f"배열 타입: {type(ndarray)}")
print(f"배열 형태: {ndarray.shape}, {ndarray.ndim}D")
print(f"배열 요소 타입: {ndarray.dtype}")
print(f"배열 전체 요소 수: {ndarray.size}")
print(f"배열 0축 요소 수: {len(ndarray)}, {ndarray.shape[0]}")
print(ndarray)
```

2차원 배열

```python
import numpy as np

ndarray = np.array([[-1, -5, 84],
                    [2, 0.5, 56],
                    [7, 1.9, 1],
                    [-19, 6, 8]])

print(f"배열 타입: {type(ndarray)}")
print(f"배열 형태: {ndarray.shape}, {ndarray.ndim}D")
print(f"배열 요소 타입: {ndarray.dtype}")
print(f"배열 전체 요소수: {ndarray.size}")
print(f"배열 0축 요소수: {len(ndarray)}, {ndarray.shape[1]}")
print(ndarray)
```

3차원 배열

```python
import numpy as np

ndarray = np.array([[[ -1,  -5, 84],
                     [  2, 0.5, 56],
                     [  7, 1.9,  1],
                     [-19,   6,  8]],

                    [[ -1,  -5, 84],
                     [  2, 0.5, 56],
                     [  7, 1.9,  1],
                     [-19,   6,  8]]])

print(f"배열 타입: {type(ndarray)}")
print(f"배열 형태: {ndarray.shape}, {ndarray.ndim}D")
print(f"배열 요소 타입: {ndarray.dtype}")
print(f"배열 전체 요소수: {ndarray.size}")
print(f"배열 0차원 요소수: {len(ndarray)}, {ndarray.shape[0]}")
print(ndarray)
```
