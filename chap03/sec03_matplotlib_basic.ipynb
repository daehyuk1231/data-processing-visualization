```python
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정 (Windows)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 음수(-) 기호가 깨지는 문제 해결
# 유니코드 마이너스 대신 ASCII 하이픈(-)을 사용
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.arange(0, 2*np.pi, 0.1)

y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure(1) # Figure1 생성 및 선택
axes = plt.gca()    # gca: get current axes
axes.set_title('사인 그래프')  # 그래프 제목 설정
axes.plot(x, y1)

fig =plt.figure(2, figsize=(6.4, 4.8))  # Figure2 생성 및 선택
axes = plt.gca()    # gca: get current axes  
axes.set_title('코사인 그래프')  # 그래프 제목 설정
axes.plot(x, y2)

plt.show()          # 그래프 표시
```
