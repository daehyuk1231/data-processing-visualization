```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정 (Windows)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 음수(-) 기호가 깨지는 문제 해결
# 유니코드 마이너스 대신 ASCII 하이픈(-)을 사용
plt.rcParams['axes.unicode_minus'] = False
```

### 판다스 시각화

```python
# 데이터프레임 생성
df = pd.DataFrame({
    "month": ["1월", "2월", "3월", "4월", "5월"],
    "product": ["A", "B", "C", "D", "E"],
    "sales": [120, 150, 170, 160, 200],
    "profit": [20, 30, 50, 40, 60]
})

# 기본 선 그래프 생성
axes = df.plot()
```

```python
# 막대 그래프 생성
axes = df.plot(kind="bar")
```

```python
# 히스토그램 생성
axes = df.plot(kind="hist") # df.hist()
```

```python
# 이상치 만들기
df.loc[0, "sales"] = 50
df.loc[1, "profit"] = 150

# 박스 플롯 생성
axes =df.plot(kind="box") # 또는 df.boxplot()
```
