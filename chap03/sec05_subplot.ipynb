```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정 (Windows)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 음수(-) 기호가 깨지는 문제 해결
# 유니코드 마이너스 대신 ASCII 하이픈(-)을 사용
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 스탭 함수
def step(x):
    return np.array(x>0, dtype=np.int32)

# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 +np.exp(-x))

# 렐루 함수
def relu(x):
    return np.maximum(0, x)

# 입력값
x = np.arange(-10, 10, 0.1)
```

```python
# 너비 6인치, 높이 6인치 Figure 생성
fig, axes = plt.subplots(3, 2, figsize=(6, 6))
print(type(axes))

# 전체 Figure 제목
fig.suptitle("서브플롯")

# 서브플로팅하기
axes[0, 0].set_title("sign 함수")
axes[0, 0].plot(x, np.sign(x))

axes[0, 1].set_title("tanh 함수")
axes[0, 1].plot(x, np.tanh(x))

axes[1, 0].set_title("step 함수")
axes[1, 0].plot(x, step(x))

axes[1, 1].set_title("sigmoid 함수")
axes[1, 1].plot(x, sigmoid(x))

axes[2, 0].set_title("relu 함수")
axes[2, 0].plot(x, relu(x))

# axes[2, 1] 제거
fig.delaxes(axes[2, 1])

# 서브플롯 간 여백 조정 (hspace: 상하 간격, wspace: 좌우 간격)
fig.subplots_adjust(hspace=0.7, wspace=0.3)

plt.show()
```

```python
# 스탭 함수
def step(x):
    return np.array(x>0, dtype=np.int32)

# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 +np.exp(-x))

# 렐루 함수
def relu(x):
    return np.maximum(0, x)

# 입력값
x = np.arange(-10, 10, 0.1)

# 너비 6인치, 높이 6인치 Figure 생성
fig = plt.figure(figsize=(6,6))

# 전체 Figure 제목
fig.suptitle("서브플롯")

# 서브플로팅하기
axes1 = plt.subplot(3,2,1)
axes1.set_title("sign 함수")
axes1.plot(x, np.sign(x))

axes2 = plt.subplot(3,2,2)
axes2.set_title("tanh 함수")
axes2.plot(x, np.tanh(x))

axes3 = plt.subplot(3,2,3)
axes3.set_title("step 함수")
axes3.plot(x, step(x))

plt.subplot(3,2,4)
plt.title("sigmoid 함수")
plt.plot(x, sigmoid(x))

plt.subplot(3,2,5)
plt.title("relu 함수")
plt.plot(x, relu(x))

# 서브플롯 간 여백 조정 (hspace: 상하 간격, wspace: 좌우 간격)
plt.subplots_adjust(hspace=0.7, wspace=0.3)

plt.show()
```
