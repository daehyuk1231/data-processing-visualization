```python
from PIL import Image
import matplotlib.pyplot as plt

file_path = "../images/photo1.jpg"
img = Image.open(file_path)

# [방법1]
plt.imshow(img)
plt.show()

# [방법2]
# plt.figure()
# axes = plt.gca()
# axes.imshow(img)
# plt.show()
```

```python
# RGB 이미지를 그레이스케일로 변환
# L: Luminance(명암) 모드, 8비트 픽셀, 흑백 이미지
img_gray = img.convert('L')

plt.imshow(img_gray, cmap="gray")
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

file_path = "../images/photo2.jpg"

# PIL ImageFile 객체로 이미지 읽기
img = Image.open(file_path)
# 이미지 크기 조정
img = img.resize((640, 480))
# PIL Image 객체를 NumPy 배열로 변환
img = np.array(img)   
print(type(img), img.shape)

# NumPy 배열로 변환된 이미지를 Matplotlib으로 시각화
plt.imshow(img)
plt.show()
```
