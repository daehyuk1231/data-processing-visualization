```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 한글 폰트 설정 (Windows)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 음수(-) 기호가 깨지는 문제 해결
# 유니코드 마이너스 대신 ASCII 하이픈(-)을 사용
plt.rcParams['axes.unicode_minus'] = False
```

##### plt.plot(): 선 그래프

```python
# 월별 매출 데이터를 DataFrame으로 생성
df = pd.DataFrame({
    "month": ["1월", "2월", "3월", "4월", "5월"],
    "sales": [120, 150, 170, 160, 200],
    "profit": [20, 30, 50, 40, 60]
})

# 선 그래프 생성
plt.figure()

axes = plt.gca()  # 현재 축 가져오기

axes.plot(df["month"], df["sales"], label="매출")
axes.plot(df["month"], df["profit"], label="이익")

axes.set_title("월별 매출 및 이익 추이") # 그래프 제목
axes.set_xlabel("월")    # X축 레이블
axes.set_ylabel("금액")  # Y축 레이블
axes.legend()   # 범례 표시

plt.show()
```

##### plt.bar(): 막대 그래프

```python
# 막대 그래프 생성
plt.figure()

axes = plt.gca()  # 현재 축 가져오기

axes.bar(df["month"], df["sales"], label="매출")
axes.bar(df["month"], df["profit"], label="이익")

axes.set_title("월별 매출 및 이익 추이") # 그래프 제목
axes.set_xlabel("월")    # X축 레이블
axes.set_ylabel("금액")  # Y축 레이블
axes.legend()        # 범례 표시

plt.show()
```

##### plt.hist(): 히스토그램

```python
# 난수 데이터 생성 (정규 분포 형태)
np.random.seed(42)
# 평균이 70이고 표준편차가 10인 정규 분포에서 100개의 점수 생성
scores = np.random.normal(loc=70, scale=10, size=100)

# 히스토그램 생성
plt.figure()

axes = plt.gca()  # 현재 축 가져오기

axes.hist(scores, bins=10)

axes.set_title("시험 점수 분포")
axes.set_xlabel("점수")
axes.set_ylabel("빈도수")

plt.show()
```

### scatter(): 산점도

```python
# 공부 시간과 시험 점수 데이터
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
scores = [50, 55, 65, 70, 75, 85, 90, 95]

# 산점도 생성
plt.figure()

axes = plt.gca()  # 현재 축 가져오기

axes.scatter(study_hours, scores)

axes.set_title("공부 시간과 시험 점수의 관계")
axes.set_xlabel("공부 시간(시간)")
axes.set_ylabel("시험 점수")

plt.show()
```
