### 데이터 요약 및 집계

```python
import pandas as pd

dataframe = pd.DataFrame({
    "employee_id": [101, 102, 103, 104, 105, 106, 107, 108, 109],
    "name": ["김민수", "박영희", "이철수", "최지영", "정선미", "홍길동",
             "강민준", "윤서연", "임재현"],
    "dept": ["개발팀", "개발팀", "분석팀", "분석팀", "인사팀", "개발팀",
             "분석팀", "인사팀", "개발팀"],
    "position": ["사원", "대리", "대리", "과장", "차장", "사원",
                 "사원", "대리", "과장"],
    "salary": [3000, 4500, 4500, 6000, 7000, 3200, 3100, 4800, 6500],
    "experience": [1, 3, 3, 5, 8, 1, 2, 4, 6]
})
dataframe
```

##### 컬럼별 집계 메소드

```python
# 최대값, 중앙값, 최소값
print("최대값:")
print(dataframe[["salary", "experience"]].max(), "\n")

print("평균값:")
print(dataframe[["salary", "experience"]].mean(), "\n")

print("중앙값:")
print(dataframe[["salary", "experience"]].median(), "\n")

print("최소값:")
print(dataframe[["salary", "experience"]].min(), "\n")

print("표준편차:")
print(dataframe[["salary", "experience"]].std())
```

```python
total_rows = len(dataframe)  # dataframe.shape[0] 도 가능
print("전체행 개수:", total_rows)

data_count = dataframe["salary"].count()
print("데이터 개수:", data_count)

missing_count = dataframe["salary"].isna().sum()
print("결측값 개수:", missing_count)
```

##### 숫자 컬럼을 대상으로 통계 얻기

```python
# 숫자형 컬럼의 기술 통계
df_result = dataframe.describe()
df_result
# print(df_result.loc["count", "salary"])
```

##### 여러 집계 함수 동시 적용

```python
# 여러 집계 함수 동시 적용
df_result = (dataframe[["salary", "experience"]]
            .agg(["mean", "sum", "min", "max"]))
df_result
```

```python
# 사용자 정의 집계 함수 추가
def peak_to_peak(series):
    return series.max() - series.min()

result = (dataframe["salary"].agg(["max", "min", peak_to_peak]))
result
```

##### 기본 그룹핑

```python
result = dataframe.groupby("dept").size()
result
```

```python
result = (dataframe
            .groupby("dept")
            .size()
            .reset_index(name="employee_count"))
result
```

```python
# 결측값 만들기
dataframe.loc[0, "experience"] = None

# 그룹핑 후 count()는 결측값을 제외한 개수를 반환
result = dataframe.groupby("dept").count()
result
```

##### 다른 컬럼을 집계하기

```python
result = (dataframe.groupby("dept")["salary"]
            .mean()
            .reset_index(name="average_salary"))
result
```

```python
result = (dataframe
            .groupby("dept")[["salary", "experience"]]
            .mean()
            .reset_index()
            .rename(columns={"salary": "average_salary", 
                             "experience": "average_experience"}))
result
```

```python
result = (dataframe
            .groupby("dept")["salary"]
            .agg(["mean", "min", "max", "sum", "count"])
            .reset_index()
            .rename(columns={"mean": "avg_sal",
                             "min": "min_sal",
                             "max": "max_sal",
                             "sum": "total_sal",
                             "count": "emp_count"}))
result
```

##### 다중 그룹핑

```python
result = (dataframe
            .groupby(["dept", "position"])
            .size()
            .reset_index(name="count"))
result
```

```python
result = (dataframe.groupby(["dept", "position"])["salary"]
            .mean()
            .reset_index()
            .rename(columns={"salary": "average_salary"}))
result
```

##### 그룹핑 후, 필터링

```python
result = dataframe.groupby("dept").filter(
    lambda dept: dept["salary"].mean() >= 5000
)
result
```

##### Pivot 테이블

```python
pivot_salary = dataframe.pivot_table(
    index="dept",              # 행 레이블이 될 컬럼
    columns="position",        # 컬럼 레이블이 될 컬럼
    values="salary",           # 집계할 컬럼
    aggfunc="mean"             # 집계 함수
).reindex(columns=["차장", "과장", "대리", "사원"]) # 컬럼 순서 재정렬
pivot_salary
```
