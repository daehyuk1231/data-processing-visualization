```python
import pandas as pd
import numpy as np
```

## 날짜/시간 처리

```python
# 샘플 데이터 생성
df_date = pd.DataFrame({
    "employee_id": [101, 102, 103],
    "name": ["김민수", "박영희", "이철수"],
    "hire_date": ["2020-01-15", "2019-06-20", "2021-03-10"],
    "birth_date": ["1990-05-20", "1985-09-15", "1992-12-03"]
})
print(df_date)
print(f"\n데이터 타입:\n{df_date.dtypes}")
```

### 문자열을 날짜로 변환

```python
# to_datetime()으로 날짜 형식 변환
df_date["hire_date"] = pd.to_datetime(df_date["hire_date"])
df_date["birth_date"] = pd.to_datetime(df_date["birth_date"])
print(df_date)
print(f"\n변환 후 데이터 타입:\n{df_date.dtypes}")
```

### 다양한 날짜 형식 파싱

```python
# 다양한 형식의 날짜 문자열
date_formats = pd.DataFrame({
    "date_str": [
        "2024/01/15",
        "15-01-2024",
        "January 15, 2024",
        "2024.01.15",
        "20240115"
    ]
})

# 자동 파싱 (각 요소마다 형식을 개별적으로 추론)
date_formats["parsed_date"] = pd.to_datetime(date_formats["date_str"], format='mixed')
print(date_formats)
```

```python
# 특정 형식 지정 (format parameter)
custom_format = pd.DataFrame({
    "date_str": ["15/01/2024", "20/06/2024", "10/12/2024"]
})

# 일/월/년 형식 명시
custom_format["date"] = pd.to_datetime(
    custom_format["date_str"], 
    format="%d/%m/%Y"
)
print(custom_format)
```

### 날짜 구성요소 추출

```python
# 연, 월, 일 추출
df_copy = df_date.copy()
df_copy["hire_year"] = df_copy["hire_date"].dt.year
df_copy["hire_month"] = df_copy["hire_date"].dt.month
df_copy["hire_day"] = df_copy["hire_date"].dt.day
print(df_copy[["name", "hire_date", "hire_year", "hire_month", "hire_day"]])
```

```python
# 요일 정보 추출
df_copy = df_date.copy()
df_copy["hire_dayofweek"] = df_copy["hire_date"].dt.dayofweek  # 0=월요일, 6=일요일
df_copy["hire_day_name"] = df_copy["hire_date"].dt.day_name()  # 요일 이름
print(df_copy[["name", "hire_date", "hire_dayofweek", "hire_day_name"]])
```

```python
# 분기 정보
df_copy = df_date.copy()
df_copy["hire_quarter"] = df_copy["hire_date"].dt.quarter
print(df_copy[["name", "hire_date", "hire_quarter"]])
```

```python
# 주차 정보
df_copy = df_date.copy()
df_copy["week_of_year"] = df_copy["hire_date"].dt.isocalendar().week
print(df_copy[["name", "hire_date", "week_of_year"]])
```

### 날짜 연산

```python
# 근속연수 계산
df_copy = df_date.copy()
today = pd.Timestamp.now()
df_copy["years_of_service"] = (today - df_copy["hire_date"]).dt.days // 365
print(df_copy[["name", "hire_date", "years_of_service"]])
```

```python
# 나이 계산
df_copy = df_date.copy()
today = pd.Timestamp.now()
df_copy["age"] = (today - df_copy["birth_date"]).dt.days // 365
print(df_copy[["name", "birth_date", "age"]])
```

```python
# 두 날짜 사이의 차이 (일수)
df_copy = df_date.copy()
df_copy["days_diff"] = (df_copy["hire_date"] - df_copy["birth_date"]).dt.days
print(df_copy[["name", "birth_date", "hire_date", "days_diff"]])
```

```python
# 날짜에 특정 기간 더하기/빼기
df_copy = df_date.copy()
df_copy["probation_end"] = df_copy["hire_date"] + pd.Timedelta(days=90)  # 수습 종료일 (90일 후)
df_copy["contract_end"] = df_copy["hire_date"] + pd.DateOffset(years=1)  # 계약 종료일 (1년 후)
print(df_copy[["name", "hire_date", "probation_end", "contract_end"]])
```

### 날짜 필터링

```python
# 2020년 이후 입사자
recent_hires = df_date[df_date["hire_date"] >= "2020-01-01"]
print(recent_hires[["name", "hire_date"]])
```

```python
# 특정 기간 범위
year_2019_2021 = df_date[
    (df_date["hire_date"] >= "2019-01-01") & 
    (df_date["hire_date"] <= "2021-12-31")
]
print(year_2019_2021[["name", "hire_date"]])
```

```python
# 특정 월에 입사한 사람
hired_in_jan = df_date[df_date["hire_date"].dt.month == 1]
print(hired_in_jan[["name", "hire_date"]])
```

```python
# 1분기에 입사한 사람
q1_hires = df_date[df_date["hire_date"].dt.quarter == 1]
print(q1_hires[["name", "hire_date"]])
```

### 날짜 포맷팅

```python
# 날짜를 다양한 형식의 문자열로 변환
df_copy = df_date.copy()
df_copy["date_format1"] = df_copy["hire_date"].dt.strftime("%Y년 %m월 %d일")
df_copy["date_format2"] = df_copy["hire_date"].dt.strftime("%Y/%m/%d")
df_copy["date_format3"] = df_copy["hire_date"].dt.strftime("%d-%b-%Y")  # 15-Jan-2020
print(df_copy[["hire_date", "date_format1", "date_format2", "date_format3"]])
```

### 날짜 범위 생성

```python
# 일별 날짜 범위
date_range = pd.date_range(start="2024-01-01", end="2024-01-10", freq="D")
print("일별 범위:")
print(date_range)
```

```python
# 월별 날짜 범위
monthly_range = pd.date_range(start="2024-01-01", periods=12, freq="MS")  # MS = Month Start
print("월별 범위:")
print(monthly_range)
```

```python
# 영업일 기준 날짜 범위 (주말 제외)
business_days = pd.bdate_range(start="2024-01-01", periods=10)
print("영업일 범위:")
print(business_days)
```

### 시계열 데이터 리샘플링

```python
# 일별 매출 데이터 생성
np.random.seed(42)
dates = pd.date_range(start="2024-01-01", end="2024-01-31", freq="D")
sales_data = pd.DataFrame({
    "date": dates,
    "sales": np.random.randint(100, 500, size=len(dates))
})
sales_data.set_index("date", inplace=True)
print("일별 매출 (처음 10일):")
print(sales_data.head(10))
```

```python
# 주별 평균 매출
weekly_avg = sales_data.resample("W").mean()
print("주별 평균 매출:")
print(weekly_avg)
```

```python
# 주별 총 매출
weekly_sum = sales_data.resample("W").sum()
print("주별 총 매출:")
print(weekly_sum)
```

### 실전 예제: 종합 데이터 전처리

```python
# 실무 예제: 고객 데이터 정제
customer_data = pd.DataFrame({
    "customer_id": ["C001", "C002", "C003"],
    "name": ["  홍길동  ", "KIM MINSOO", "park younghee"],
    "email": ["hong@EXAMPLE.com", "kim@test.COM", "park@example.org"],
    "phone": ["010-1234-5678", "01098765432", "010-5555-6666"],
    "registration_date": ["2023-01-15", "2023-06-20", "2024-01-10"]
})

print("원본 데이터:")
print(customer_data)
```

```python
# 데이터 정제 파이프라인
cleaned_data = customer_data.copy()

# 1. 이름 정제: 공백 제거, 타이틀 케이스
cleaned_data["name"] = cleaned_data["name"].str.strip().str.title()

# 2. 이메일 정제: 소문자 변환
cleaned_data["email"] = cleaned_data["email"].str.lower()

# 3. 전화번호 정제: 하이픈 형식 통일 (010-XXXX-XXXX)
cleaned_data["phone"] = cleaned_data["phone"].str.replace(r'[\s-]', '', regex=True)  # 공백/하이픈 제거
cleaned_data["phone"] = cleaned_data["phone"].str.replace(
    r'(\d{3})(\d{4})(\d{4})', 
    r'\1-\2-\3', 
    regex=True
)  # 하이픈 추가

# 4. 날짜 변환 및 가입 기간 계산
cleaned_data["registration_date"] = pd.to_datetime(cleaned_data["registration_date"])
cleaned_data["days_since_registration"] = (pd.Timestamp.now() - cleaned_data["registration_date"]).dt.days

# 5. 이메일 도메인 추출
cleaned_data["email_domain"] = cleaned_data["email"].str.split("@").str[1]

print("\n정제된 데이터:")
print(cleaned_data)
```

```python
# 분석: 도메인별 고객 수
domain_counts = cleaned_data["email_domain"].value_counts()
print("\n도메인별 고객 수:")
print(domain_counts)
```

```python
# 분석: 월별 가입 현황
cleaned_data["registration_month"] = cleaned_data["registration_date"].dt.to_period("M")
monthly_registrations = cleaned_data.groupby("registration_month").size()
print("\n월별 가입자 수:")
print(monthly_registrations)
```
