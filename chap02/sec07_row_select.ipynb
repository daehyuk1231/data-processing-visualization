### 데이터 검색

```python
import pandas as pd

dataframe = pd.DataFrame(data={
    "eno": [100, 200, 300, 400, 500],
    "name": ["춘봄이", "하여름", "추가을", "동겨울", "춘순이"],
    "dept": ["개발팀", "분석팀", "홍보팀", "인사팀", "설계팀"],
    "date": [
        pd.to_datetime("2015-09-22"),
        pd.to_datetime("2016-09-22"),
        pd.to_datetime("2017-09-22"),
        pd.to_datetime("2018-09-22"),
        pd.to_datetime("2019-09-22")
    ],
    "salary": [8000, 7500, 7300, 7200, 7700]
})

dataframe
```

```python

dataframe.set_index("eno", inplace=True)
dataframe
```

### 무작위 행 가져오기

```python
resultDF = dataframe.sample(3)
resultDF
```

```python
# dataframe에서 30%의 데이터를 랜덤하게 추출하여 resultDF에 저장한다.
resultDF = dataframe.sample(frac=0.3)
resultDF
```

```python
# random_state 매개변수에 1을 입력하여 난수 시드를 고정한다.
resultDF = dataframe.sample(frac=0.3, random_state=1)
resultDF
```

정렬해서 가져오기

```python
resultDF = dataframe.sort_values("salary", ascending=False)
resultDF
```
