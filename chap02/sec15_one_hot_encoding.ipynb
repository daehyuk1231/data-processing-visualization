```python
import pandas as pd

dataframe = pd.DataFrame({
    "height": [175, 162, 180, 158, 170],
    "sex": ['male', 'female', 'male', 'female', 'male'],
    "nation": ['korea', 'usa', 'Germany', 'korea', 'usa']
})
dataframe
```

```python
df_copy = dataframe.copy()
# 원 핫 인코딩 컬럼 생성
dummies = pd.get_dummies(df_copy[["sex", "nation"]], dtype=int)
# 컬럼 방향으로 결합하고 기존 컬럼 제거
df_result = (pd.concat([df_copy, dummies], axis=1)
              .drop(columns=["sex", "nation"]))
df_result
```
