### 데이터 연산

전체 스칼라 연산

```python
import numpy as np
import pandas as pd

dataframe = pd.DataFrame(np.ones((3, 3)), columns=["A", "B", "C"])
dataframe += 10
dataframe
```

컬럼별 스칼라 연산

```python
dataframe = pd.DataFrame(np.ones((3, 3)), columns=["A", "B", "C"])
dataframe[["B", "C"]] += [10, -10]
dataframe
```

슬라이싱 범위 연산

```python
dataframe = pd.DataFrame(np.ones((3, 3)), columns=["A", "B", "C"])
dataframe.loc[0:1,"B":"C"] += 10
dataframe
```

```python
dataframe = pd.DataFrame(np.ones((3, 3)), columns=["A", "B", "C"])
dataframe.loc[:, "B":"C"] += [10, 10]
dataframe
```
