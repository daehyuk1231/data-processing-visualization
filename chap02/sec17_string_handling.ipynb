```python
import pandas as pd
```

```python
# 샘플 데이터 생성
dataframe = pd.DataFrame({
    "name": ["  Kim MinSu  ", "Park YoungHee", "Lee ChulSoo"],
    "email": ["minsu@example.com", "younghee@test.com", "chulsoo@example.org"],
    "phone": ["010-1234-5678", "010-9876-5432", "011-1111-2222"],
    "address": ["서울시 강남구", "경기도 수원시", "서울시 송파구"]
})
print(dataframe)
```

### 대소문자 변환

```python
# 소문자로 변환
df_copy = dataframe.copy()
df_copy["name_lower"] = df_copy["name"].str.lower()
print(df_copy[["name", "name_lower"]])
```

```python
# 대문자로 변환
df_copy = dataframe.copy()
df_copy["name_upper"] = df_copy["name"].str.upper()
print(df_copy[["name", "name_upper"]])
```

```python
# 첫 글자만 대문자로 (Title Case)
df_copy = dataframe.copy()
df_copy["name_title"] = df_copy["name"].str.title()
print(df_copy[["name", "name_title"]])
```

### 공백 제거

```python
# 양쪽 공백 제거
df_copy = dataframe.copy()
df_copy["name_strip"] = df_copy["name"].str.strip()
print(df_copy[["name", "name_strip"]])
print(f"\n원본 길이: {len(df_copy['name'].loc[0])}")
print(f"제거 후 길이: {len(df_copy['name_strip'].loc[0])}")
```

```python
# 왼쪽 공백만 제거
print(dataframe["name"].str.lstrip())
print()

# 오른쪽 공백만 제거
print(dataframe["name"].str.rstrip())
```

### 문자열 포함 여부 확인

```python
# 특정 문자열 포함 여부
df_copy = dataframe.copy()
df_copy["has_kim"] = df_copy["name"].str.contains("Kim", case=True)
print(df_copy[["name", "has_kim"]])
```

```python
# 대소문자 구분 없이 검색
df_copy = dataframe.copy()
df_copy["has_kim"] = df_copy["name"].str.contains("kim", case=False)
print(df_copy[["name", "has_kim"]])
```

```python
# 서울시에 거주하는 사람 필터링
seoul_residents = dataframe[dataframe["address"].str.contains("서울시")]
print(seoul_residents)
```

```python
# example.com 이메일 사용자 필터링
example_users = dataframe[dataframe["email"].str.contains("example.com")]
print(example_users[["name", "email"]])
```

### 문자열 시작/끝 확인

```python
# 010으로 시작하는 전화번호
modern_phones = dataframe[dataframe["phone"].str.startswith("010")]
print(modern_phones[["name", "phone"]])
```

```python
# .com으로 끝나는 이메일
com_emails = dataframe[dataframe["email"].str.endswith(".com")]
print(com_emails[["name", "email"]])
```

### 문자열 분리 (split)

```python
# 이메일에서 사용자명과 도메인 분리
df_copy = dataframe.copy()

# str.split() 메서드를 사용하여 "@"를 기준으로 이메일을 분리하고, 
# expand=True로 결과를 DataFrame 형태로 반환
email_parts = df_copy["email"].str.split("@", expand=True)
print(email_parts, "\n")

# 분리된 결과를 새로운 열로 추가
df_copy["username"] = email_parts[0]
df_copy["domain"] = email_parts[1]

# 결과 출력
print(df_copy[["email", "username", "domain"]])
```

```python
# 전화번호 분리
df_copy = dataframe.copy()
phone_parts = df_copy["phone"].str.split("-", expand=True)
df_copy["phone_1"] = phone_parts[0]
df_copy["phone_2"] = phone_parts[1]
df_copy["phone_3"] = phone_parts[2]
print(df_copy[["phone", "phone_1", "phone_2", "phone_3"]])
```

```python
# 주소에서 시/도 추출
df_copy = dataframe.copy()
df_copy["city"] = df_copy["address"].str.split().str[0]
df_copy["district"] = df_copy["address"].str.split().str[1]
print(df_copy[["address", "city", "district"]])
```

### 문자열 대체 (replace)

```python
# 하이픈 제거
df_copy = dataframe.copy()
df_copy["phone_no_dash"] = df_copy["phone"].str.replace("-", "")
print(df_copy[["phone", "phone_no_dash"]])
```

```python
# 특정 단어 대체
df_copy = dataframe.copy()
df_copy["address_short"] = df_copy["address"].str.replace("서울시", "서울")
print(df_copy[["address", "address_short"]])
```

### 문자열 길이

```python
# 문자열 길이 확인
df_copy = dataframe.copy()
df_copy["name_length"] = df_copy["name"].str.len()
df_copy["email_length"] = df_copy["email"].str.len()
print(df_copy[["name", "name_length", "email", "email_length"]])
```

### 문자열 추출 (슬라이싱)

```python
# 전화번호 앞 3자리 추출
df_copy = dataframe.copy()
df_copy["phone_prefix"] = df_copy["phone"].str[:3]
print(df_copy[["phone", "phone_prefix"]])
```
