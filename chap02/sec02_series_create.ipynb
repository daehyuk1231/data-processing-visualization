```python
import pandas as pd
import numpy as np
```

```python
series = pd.Series(["오렌지", "바나나", "딸기"])
print(series)
```

```python
series = pd.Series(
    data=np.array(["오렌지", "바나나", "딸기", "사과"]),
    index=["F1", "F2", "F3", "F4"]
)

print(series)
```

```python
series = pd.Series(
    data=np.array(["오렌지", "바나나", "딸기", "사과"]),
    index=["F1", "F2", "F3", "F4"],
    name="과일"
)

print(series)
```
