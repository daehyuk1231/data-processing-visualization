```python
import pandas as pd

dataframe = pd.DataFrame(data={
    "eno": [1, 2, 3, 4, 5, 6, 7, 8],
    "name": ["춘봄", "하름", "추을", "동울", "춘이", "홍동", "홍서", "홍남"],
    "dept": ["개발", "분석", "홍보", "인사", "설계", "개발", "개발", "분석"],
    "date": [
        pd.to_datetime("2015-09-22"),
        pd.to_datetime("2016-09-22"),
        pd.to_datetime("2017-09-22"),
        pd.to_datetime("2018-09-22"),
        pd.to_datetime("2019-09-22"),
        pd.to_datetime("2020-09-22"),
        pd.to_datetime("2021-09-22"),
        pd.to_datetime("2021-09-22")
    ],
    "salary": [8000, 7500, 7300, 7200, 7700, 7800, 7900, 7600]
})

dataframe
```

```python
# 레이블로 단일 행 삭제
dataframe.drop(1, inplace=True)
dataframe
```

```python
# 여러 행 삭제
dataframe.drop([3, 4], inplace=True)
dataframe
```

```python
# 레이블 범위로 여러 행 삭제
# dataframe.loc[2:5]는 양끝을 포함하는 범위로 행을 선택 
# dataframe.loc[2:5].index는 행 Index를 반환
dataframe.drop(dataframe.loc[2:5].index, inplace=True)
dataframe
```
