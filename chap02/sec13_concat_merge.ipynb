### DataFrame 결합 및 병합

```python
import pandas as pd
```

##### 행 방향으로 결합

```python
df1 = pd.DataFrame({
    "name": ["홍길동", "김영수"],
    "age": [25, 30],
    "city": ["서울", "부산"]
})

df2 = pd.DataFrame({
    "name": ["이철수", "박민지"],
    "age": [28, 35],
    "city": ["대구", "인천"]
})

result = pd.concat([df1, df2], ignore_index=True)
result
```

##### 컬럼 방향으로 결합

```python
df1 = pd.DataFrame({
    "name": ["김철수", "박영희", "이민수"],
    "dept": ["개발팀", "분석팀", "인사팀"]
})

df2 = pd.DataFrame({
    "salary": [8000, 7500, 7300],
    "bonus": [500, 300, 400]
})

result = pd.concat([df1, df2], axis=1)
result
```

##### 병합: Inner Join

```python
employees = pd.DataFrame({
    "emp_id": [1, 2, 3, 4],
    "name": ["김철수", "박영희", "이민수", "최지영"],
    "dept_id": [101, 102, 101, None]
})

departments = pd.DataFrame({
    "dept_id": [101, 102, 103, 104],
    "dept_name": ["개발팀", "분석팀", "인사팀", "영업팀"]
})

print("직원 정보:")
print(employees)

print("\n부서 정보:")
print(departments)

result = pd.merge(employees, departments, on="dept_id", how="inner")
print("\nInner Join 결과:")
print(result)
```

##### 병합: Left Join

```python
# Left Join: 왼쪽(employees) DataFrame의 모든 행 유지
result = pd.merge(employees, departments, on="dept_id", how="left")
result
```

##### 병합: Right Join

```python
# Right Join: 오른쪽(departments) DataFrame의 모든 행 유지
result = pd.merge(employees, departments, on="dept_id", how="right")
result
```

##### 병합: Outer Join

```python
# Outer Join: 양쪽 모든 행 유지 (결측값은 NaN)
result = pd.merge(employees, departments, on="dept_id", how="outer")
result
```

##### 주문 데이터와 고객 데이터 병합

```python
# 주문 데이터
orders = pd.DataFrame({
    "order_id": [1001, 1002, 1003, 1004],
    "customer_id": ["C001", "C002", "C001", "C003"],
    "product": ["노트북", "마우스", "키보드", "모니터"],
    "price": [1200000, 30000, 80000, 350000]
})
orders
```

```python
# 고객 데이터
customers = pd.DataFrame({
    "customer_id": ["C001", "C002", "C003"],
    "customer_name": ["김철수", "박영희", "이민수"],
    "city": ["서울", "부산", "대구"]
})
customers
```

```python

# DataFrame 병합 (고객 정보와 주문 정보 결합)
merged_df = pd.merge(orders, customers, on="customer_id", how="inner")
merged_df
```

```python
# 고객별 총 구매 금액과 구매 횟수 집계
result = (merged_df.groupby(["customer_id", "customer_name"])["price"]
                   .agg(["sum", "count"])
                   .reset_index()
                   .rename(columns={"sum": "total_price", 
                                    "count": "purchase_count"}))
result
```
