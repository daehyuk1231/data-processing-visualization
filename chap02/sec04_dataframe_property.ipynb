### DataFrame 속성

```python
import pandas as pd
import numpy as np
```

DataFrame 생성

```python
pname = ["오렌지", "바나나", "딸기", "사과"]
pprice = [3000, 2000, 6000, 4500]
pstock = [50, 30, 10, 100]
porigin = ["미국", "필리핀", "한국", "한국"]
pdate = ["2019-01-01", "2019-02-01", "2019-03-01", "2019-04-01"]

dataframe = pd.DataFrame({
    "PNAME":pname,
    "PPRICE":pprice,
    "PSTOCK":pstock,
    "PORIGIN":porigin,
    "PDATE": pdate
})

dataframe

```

index 속성과 columns 속성 

```python
print(dataframe.index)
print(dataframe.columns)
```

```python
dataframe.index = ["F1", "F2", "F3", "F4"]
dataframe.columns = ["상품명", "가격", "재고", "원산지", "수확일"]
dataframe
```

### dtypes 속성

```python
dataframe.dtypes
```

```python
dataframe["가격"] = dataframe["가격"].astype("float")
dataframe.dtypes
```

```python
new_dataframe = dataframe.astype({
    "가격": "float",
    "재고": "int",
    "수확일": "datetime64[s]"
})

new_dataframe.dtypes
```

values 속성

```python
# 둘다 반환 타입이 ndarray
values = dataframe.values
# values = dataframe.to_numpy()

print(type(values), "\n")
print(values)
```

```python
# 반환 타입이 StringArray
# values = dataframe["상품명"].values

# 반환 타입이 ndarray
values = dataframe["상품명"].to_numpy()

print(type(values), "\n")
print(values)
```
