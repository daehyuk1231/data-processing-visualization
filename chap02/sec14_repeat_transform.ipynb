### 반복 및 변환

```python
import pandas as pd
```

```python
# 샘플 데이터 준비
dataframe = pd.DataFrame({
    "eno": [1, 2, 3, 4, 5],
    "name": ["춘봄이", "하여름", "추가을", "춘순이", "김철수"],
    "dept": ["개발팀", "분석팀", "홍보팀", "설계팀", "개발팀"],
    "date": pd.to_datetime([
        "2015-09-22", "2016-09-22", "2017-09-22",
        "2019-09-22", "2020-03-15"
    ]),
    "salary": [8000, 7500, 7300, 7700, 7100],
    "bonus": [100, 200, 300, 500, 250]
})

dataframe
```

##### itertuples() - 행 단위 반복

```python
df_copy = dataframe.copy()
grade_list = []
for row in df_copy.itertuples():
    if row.salary >= 8000 and row.bonus >= 500:
        grade = "S"
    elif row.salary >= 7500 or row.bonus >= 400:
        grade = "A"
    elif row.salary >= 7000:
        grade = "B"
    else:
        grade = "C"    
    grade_list.append(grade)
df_copy["grade"] = grade_list
df_copy
```

##### 컬럼 변환

```python
# Min-Max 정규화 적용하기
df_copy = dataframe.copy()
df_copy["salary"] = df_copy["salary"].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
df_copy
```

```python
# 여러 컬럼에 동시에 적용하기
df_copy = dataframe.copy()
df_copy[["salary", "bonus"]] = df_copy[["salary", "bonus"]].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
df_copy
```

```python
# 표준화(평균=0, 표준편차=1) 적용하기
df_copy = dataframe.copy()
df_copy[["salary", "bonus"]] = df_copy[["salary", "bonus"]].transform(
    lambda col: (col - col.mean()) / col.std()
)
df_copy
```

##### 행 변환

```python
df_copy = dataframe.copy()
df_copy["total_pay"] = df_copy.apply(
    lambda row: row['salary'] + row['bonus'],
    axis=1
)
df_copy
```

```python
def determine_grade(row):
    total = row['salary'] + row['bonus']    
    if total >= 8500:
        return "S"
    elif total >= 7800:
        return "A"
    elif total >= 7300:
        return "B"
    else:
        return "C"

df_copy = dataframe.copy()
df_copy["grade"] = df_copy.apply(
    determine_grade, 
    axis=1
)
df_copy
```

##### 요소 변환

```python
# 딕셔너리로 값 매핑 (가장 많이 사용!)
dept_mapping = {
    "개발팀": "DEV",
    "분석팀": "DATA",
    "홍보팀": "PR",
    "인사팀": "HR",
    "설계팀": "DESIGN"
}

df_copy = dataframe.copy()
df_copy["dept"] = df_copy["dept"].map(dept_mapping)
df_copy
```
