```python
# 필요한 라이브러리 임포트
import cv2
import matplotlib.pyplot as plt

# 이미지 파일 경로 설정 및 이미지 로드
filepath = "../images/person_and_dog.jpg"
img = cv2.imread(filepath)

# 선 그리기: 시작점(20,60) → 끝점(350,60), 마젠타색, 두께 2px
img = cv2.line(img, (20,60), (350,60), (255,0,255), 2, cv2.LINE_AA)

# 첫 번째 사각형의 모서리 표시 (노란색 원, 테두리만)
img = cv2.circle(img, (80, 120), 10, (255, 255, 0), 2, cv2.LINE_AA)
img = cv2.circle(img, (380, 430), 10, (255, 255, 0), 2, cv2.LINE_AA)
# 노란색 테두리 사각형 그리기 (두께 2px)
img = cv2.rectangle(img, (80, 120), (380, 430), (255, 255, 0), 2)

# 두 번째 사각형의 모서리 표시 (흰색 원, 채우기)
img = cv2.circle(img, (400, 120), 10, (255, 255, 255), -1, cv2.LINE_AA)
img = cv2.circle(img, (630, 430), 10, (255, 255, 255), -1, cv2.LINE_AA)
# 흰색 테두리 사각형 그리기 (두께 2px)
img = cv2.rectangle(img, (400, 120), (630, 430), (255, 255, 255), 2)

# 이미지에 텍스트 추가: "Person & Dog" (노란색, 크기 1.5)
img = cv2.putText(img, "Person & Dog", (30,50),
                  cv2.FONT_HERSHEY_SCRIPT_COMPLEX,
                  1.5, (0,255,255), 2, cv2.LINE_AA)

# 첫 번째 사각형 영역의 밝기를 줄여서 어둡게 만들기 (각 채널을 1.5로 나눔)
img[120:430, 80:380, 0] = img[120:430, 80:380, 0]/1.5  # B 채널
img[120:430, 80:380, 1] = img[120:430, 80:380, 1]/1.5  # G 채널
img[120:430, 80:380, 2] = img[120:430, 80:380, 2]/1.5  # R 채널

# BGR을 RGB로 변환 (OpenCV는 BGR, matplotlib은 RGB 사용)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# matplotlib으로 시각화
plt.figure(figsize=(10, 8))
plt.imshow(img_rgb)
plt.title("Person & Dog")
plt.axis("off")
plt.show()
```
