```python
import cv2
import os

# 이미지 읽기
image = cv2.imread("../images/photo1.jpg")

# 저장 폴더 생성
os.makedirs("output", exist_ok=True)

# JPG 형식으로 저장
jpg_path = "output/created.jpg"
cv2.imwrite(jpg_path, image)

png_path = "output/created.png"
cv2.imwrite(png_path, image)

# 파일 크기 비교
jpg_size = os.path.getsize(jpg_path)
png_size = os.path.getsize(png_path)

print(f"JPG 파일 크기: {jpg_size} bytes")
print(f"PNG 파일 크기: {png_size} bytes")
```
