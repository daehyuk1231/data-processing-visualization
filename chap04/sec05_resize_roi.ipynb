```python
import cv2
import os

# 이미지 읽기
image = cv2.imread("../images/person1.jpg")
print("원본 shape:", image.shape)

# 저장 폴더 생성
os.makedirs("output", exist_ok=True)

# 고정 크기로 리사이즈 (300x300)
resized_fixed = cv2.resize(image, (300, 300))
print("고정 크기 resize:", resized_fixed.shape)


# 종횡비 유지하며 너비를 400으로 조정
height, width = image.shape[:2]
new_width = 400
new_height = int(height * (new_width / width))

resized_ratio = cv2.resize(image, (new_width, new_height))
print("종횡비 유지 resize:", resized_ratio.shape)

# ROI 추출 (좌상단: (350, 30), 우하단: (700, 450))
roi = image[30:450, 350:700]
print("ROI shape:", roi.shape)

# 결과 저장
cv2.imwrite("output/resized_fixed.jpg", resized_fixed)
cv2.imwrite("output/resized_ratio.jpg", resized_ratio)
cv2.imwrite("output/roi.jpg", roi)
```
